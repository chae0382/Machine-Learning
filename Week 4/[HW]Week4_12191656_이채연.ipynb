```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
K = 10
rand_data = np.random.rand(2,100)
x1 = np.random.rand(2,1)
x2 = np.random.rand(2,1)
x3 = np.random.rand(2,1)
```

```python
dataLen = rand_data.shape[1]

if K < dataLen:
    K = K
else:
    K = dataLen
```

### 유클리디언 거리(Euclidean distance)를 거리 함수로 KNN 알고리즘에 적용

```python
distance1 = np.sqrt(((rand_data-x1)**2).sum(axis=0))
distance2 = np.sqrt(((rand_data-x2)**2).sum(axis=0))
distance3 = np.sqrt(((rand_data-x3)**2).sum(axis=0))
# Sort the distances
idx1 = np.argsort(distance1)
idx2 = np.argsort(distance2)
idx3 = np.argsort(distance3)
# return the indices of K nearest neighbor
neighborIndex1 = idx1[:K]
neighborIndex2 = idx2[:K]
neighborIndex3 = idx3[:K]
```

```python
plt.plot(rand_data[0, :], rand_data[1, :], "ok", label="Data")
plt.plot(x1[0,0], x1[1,0], 'or', label='Query')
plt.plot(rand_data[0, neighborIndex1], rand_data[1, neighborIndex1], 'o',
        markerfacecolor='None', markersize=12,
        markeredgewidth = 0.5, label = 'KNN')
plt.legend(loc='upper right')
plt.title('KNN search(Euclidean distance) - x1')
plt.show()
```

```python
plt.plot(rand_data[0, :], rand_data[1, :], "ok", label="Data")
plt.plot(x2[0,0], x2[1,0], 'or', label='Query')
plt.plot(rand_data[0, neighborIndex2], rand_data[1, neighborIndex2], 'o',
        markerfacecolor='None', markersize=12,
        markeredgewidth = 0.5, label = 'KNN')
plt.legend(loc='upper right')
plt.title('KNN search(Euclidean distance) - x2')
plt.show()
```

```python
plt.plot(rand_data[0, :], rand_data[1, :], "ok", label="Data")
plt.plot(x3[0,0], x3[1,0], 'or', label='Query')
plt.plot(rand_data[0, neighborIndex3], rand_data[1, neighborIndex3], 'o',
        markerfacecolor='None', markersize=12,
        markeredgewidth = 0.5, label = 'KNN')
plt.legend(loc='upper right')
plt.title('KNN search(Euclidean distance) - x3')
plt.show()
```

### 맨해튼 거리(Manhattan distance)를 거리 함수로 KNN 알고리즘에 적용

```python
distance1 = abs(rand_data-x1).sum(axis=0)
distance2 = abs(rand_data-x2).sum(axis=0)
distance3 = abs(rand_data-x3).sum(axis=0)
# Sort the distances
idx1 = np.argsort(distance1)
idx2 = np.argsort(distance2)
idx3 = np.argsort(distance3)
# return the indices of K nearest neighbor
neighborIndex1 = idx1[:K]
neighborIndex2 = idx2[:K]
neighborIndex3 = idx3[:K]
```

```python
plt.plot(rand_data[0, :], rand_data[1, :], "ok", label="Data")
plt.plot(x1[0,0], x1[1,0], 'or', label='Query')
plt.plot(rand_data[0, neighborIndex1], rand_data[1, neighborIndex1], 'o',
        markerfacecolor='None', markersize=12,
        markeredgewidth = 0.5, label = 'KNN')
plt.legend(loc='upper right')
plt.title('KNN search(Manhattan distance) - x1')
plt.show()
```

```python
plt.plot(rand_data[0, :], rand_data[1, :], "ok", label="Data")
plt.plot(x2[0,0], x2[1,0], 'or', label='Query')
plt.plot(rand_data[0, neighborIndex2], rand_data[1, neighborIndex2], 'o',
        markerfacecolor='None', markersize=12,
        markeredgewidth = 0.5, label = 'KNN')
plt.legend(loc='upper right')
plt.title('KNN search(Manhattan distance) - x2')
plt.show()
```

### 코사인 거리(Cosine distance)를 거리 함수로 KNN 알고리즘에 적용

```python
l2_x1 = np.sqrt(sum(np.square(x1)))
l2_rand_data = np.sqrt(sum(np.square(rand_data)))
distance1 = 1 - np.dot(x1.T.reshape(2,), rand_data)/(l2_x1 * l2_rand_data)

l2_x2 = np.sqrt(sum(np.square(x2)))
distance2 = 1 - np.dot(x2.T.reshape(2,), rand_data)/(l2_x2 * l2_rand_data)

l2_x3 = np.sqrt(sum(np.square(x3)))
distance3 = 1 - np.dot(x3.T.reshape(2,), rand_data)/(l2_x3 * l2_rand_data)

# Sort the distances
idx1 = np.argsort(distance1)
idx2 = np.argsort(distance2)
idx3 = np.argsort(distance3)
# return the indices of K nearest neighbor
neighborIndex1 = idx1[:K]
neighborIndex2 = idx2[:K]
neighborIndex3 = idx3[:K]
```

```python
plt.plot(rand_data[0, :], rand_data[1, :], "ok", label="Data")
plt.plot(x1[0,0], x1[1,0], 'or', label='Query')
plt.plot(rand_data[0, neighborIndex1], rand_data[1, neighborIndex1], 'o',
        markerfacecolor='None', markersize=12,
        markeredgewidth = 0.5, label = 'KNN')
plt.legend(loc='upper right')
plt.title('KNN search(Cosine distance) - x1')
plt.show()
```
